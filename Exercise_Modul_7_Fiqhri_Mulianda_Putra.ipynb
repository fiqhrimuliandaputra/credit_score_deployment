```python
# TUJUAN WOE:

  # woe dibutuhkan saat logreg, misalnya max strong assumption agar monotonic

  # woe cukup powerful untuk outlier, ketika ada outlier dimasukin ke model bisa mempengaruhi
    # tapi dengan adanya woe, si outlier bisa masuk ke bin

  # woe cukup powerful dengan masalah nan
  # ketika imputasi nan, kita isi dgn majority. di mana realitasnya tidak sesuai dengan faktanya
  # di woe ini, kita melakukan binning supaya nan ini keep jadi 1 value tersendiri
  # misal: oh ternyata di umur, orang2 yg tidak ada data umurnya, woe-nya negatif. berarti orang2 yg nan ini tidak survived.
```

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score

def print_evaluate(true, predicted):
    auc = metrics.roc_auc_score(true, predicted)
    accuracy = metrics.accuracy_score(true, predicted)
    print('Accuracy:', accuracy)
    print('AUC:', auc)
    print('__________________________________')
    
def evaluate(true, predicted):
    auc = metrics.roc_auc_score(true, predicted)
    accuracy = metrics.accuracy_score(true, predicted)
    return accuracy, auc
```
